```python
import numpy as np
import matplotlib.pyplot as plt
from vmapplet.organs.leaf import AppleLeaf
from vmapplet.organs.fruit import AppleFruit
```

## Growth of tree organs over time

```python
l = AppleLeaf()
days = np.arange(0,12, 0.1)

s = []
for d in days:
    l.age = d
    # 16 so that number is greater than preformed leaves maximum value of 8.
    s.append(l.compute_area(16))

plt.plot(days, s, '-')
plt.xlabel('Time (days)')
plt.ylabel('Surface (cm^2)')
plt.grid()
_ = plt.title('Surface area of a leaf')
```

```python
l = AppleLeaf()
days = np.arange(0,12, 0.1)

m = []
for d in days:
    l.age = d
    # 16 so that number is greater than preformed leaves maximum value of 8.
    s = l.compute_area(16)
    m.append(l.compute_mass())

plt.plot(days, m, '-')
plt.xlabel('Time (days)')
plt.ylabel('Mass g')
plt.grid()
plt.title('Mass of a leaf')
```

```python
f = AppleFruit()
mass = []
age = np.arange(1, 147)
for t in age:
    f.age = t
    mass.append(f.compute_mass() * 1000)

plt.plot(age, mass)
plt.xlabel('Times(days)')
plt.ylabel('Fruit mass(grams)')
plt.grid()
_ = plt.title('Mass of a fruit')
```
